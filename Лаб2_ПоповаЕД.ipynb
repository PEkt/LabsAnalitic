**2.1.7** Напишите программу на Python для построения двух или более линий с
разными стилями. Фрагмент кода дает вывод, показанный на следующем
снимке экрана:

```
import matplotlib.pyplot as plt
y1 = [20, 40, 10]
y2 = [40, 10, 25]
x = [10.0, 20.0, 30.0]
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
line1 = ax.plot(x, y1, 'b:')
line2 = ax.plot(x, y2, 'r--')
ax.legend(labels = ('line1 - dotted', 'line2 - dashed'), loc = 'upper right')
ax.set_title('Линии с разными стилями')
ax.set_xlabel('ось x')
ax.set_ylabel('ось y')
plt.show()
```

**2.2.9** Напишите программирование на Python для отображения гистограммы
популярности языков программирования. Увеличьте нижнее поле.
Пример данных:
Языки программирования: Java, Python, PHP, JavaScript, C #, C ++
Популярность: 22.2, 17.6, 8.8, 8, 7.7, 6.7
Фрагмент кода дает вывод, показанный на следующем снимке экрана:

```
import numpy as np
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
fig.set_figheight(3)
fig.set_figwidth(7)
langs = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
students = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
ax.bar(langs,students)
ax.minorticks_on()
ax.grid(which='major',
        color = 'r', 
        linewidth = 1)
ax.grid(which='minor', 
        color = 'b', 
        linestyle = ':')       
ax.set_title('Popularity of Programming Language \nWorldwide, Oct 2017 compared to a year ago')
ax.set_xlabel('Languages')
ax.set_ylabel('Popularity')
plt.setp(ax.get_xticklabels(), rotation=90)
ax.set_xlim(-0.40, 5.6)
ax.set_ylim(0, 25)
ax.set_xticks(np.arange(-0.40, 5.6))
plt.show() 
```

**2.4.6** Напишите программу на Python, чтобы нарисовать точечный график
для трех разных групп с разной массой и высотой. 
Фрагмент кода дает вывод, показанный на следующем снимке экрана:

```
import numpy as np
import matplotlib.pyplot as plt
X = [56, 72, 62, 66, 74, 67, 59, 68]
Y = [127, 214, 256, 95, 160, 143, 85, 225]
X1 = [58, 76, 61, 65, 70, 64, 57, 70]
Y1 = [125, 202, 253, 175, 256, 140, 88, 180]
X2 = [59, 76, 61, 69, 72, 68, 58, 67]
Y2 = [250, 88, 200, 95, 253, 93, 125, 243]
plt.scatter(X, Y, s=30, c='green', marker='*')
plt.scatter(X1, Y1, s=30, c='blue', marker='*')
plt.scatter(X2, Y2, s=30, c='red', marker='*')
plt.title('Group wise Weight VS Height scatter plot')
plt.xlabel('weight')
plt.ylabel('height')
plt.xlim([55, 75])
plt.ylim([80, 260])
plt.show()
```
