```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
df=pd.read_csv('london_merged.csv')
df.drop(['timestamp','cnt','wind_speed','t1','t2'],axis=1,inplace=True)
df.head()
```

# **BOX PLOT**

```python
sns.boxplot(df['hum'])
```

```python
sns.boxplot(x='hum', data=df)
```

# **VIOLIN PLOT**

```python
sns.violinplot(x='season', y='hum', data = df)
```

# **COUNT PLOT**

```python
sns.countplot(df['is_holiday'])
```

```python
sns.countplot(df['is_weekend'])
```

# **HEATMAP**

```python
sns.heatmap(df.corr(), annot = True, cmap = 'YlGnBu')
```

# **FacetGrid**

```python
g = sns.FacetGrid(df, col="is_weekend", row="weather_code")
g.map(plt.hist,'season')
```

# **Regplot**

```python
sns.set_palette("rocket")
sns.regplot(x = 'weather_code', y = 'hum', data = df, marker="*")
```

# **Implot**

```python
sns.lmplot(x='hum', y='weather_code', row='is_weekend', data=df, col='season')
sns.set_palette("viridis")
```

# **Swarmplot**

```python
sns.swarmplot(x = 'season', y = 'hum', data = df)
sns.set_palette("pastel")
```

# **Pair plot**

```python
sns.pairplot(df, hue='hum', palette = 'magma')
```

# **Joint plot**

```python
sns.set_palette("Spectral")
sns.jointplot(x = 'season', y = 'hum', data = df)
```
