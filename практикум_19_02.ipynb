```
from matplotlib import pyplot as plt
import numpy as np
import math
x = np.arange(0, np.pi*2, 0.05)
y = np.sin(x)
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)
ax.set_title('Тригонометрическая функция')
ax.set_xlabel('угол')
ax.set_ylabel('sin()')
plt.show()
```

Задание 7

```
import matplotlib.pyplot as plt
import numpy as np
import math
x = np.arange(0, math.pi * 2, 0.05)
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.55, 0.55, 0.3, 0.3])
y = np.sin(x)
axes1.plot(x, y, 'b')
axes2.plot(x, np.cos(x), 'r')
axes1.set_title('sin')
axes2.set_title("cosin")
plt.show()
```

```
import matplotlib.pyplot as plt
fis,a = plt.subplots(2, 2)
import numpy as np
x = np.arange(1, 5)
a [0] [0].plot(x, x * x)
a [0] [0].set_title('square')
a [0] [1].plot(x, np.sqrt(x))
a [0] [1].set_title('square root')
a [1] [0].plot(x, np.exp(x))
a [1] [0].set_title('exp')
a [1] [1].plot(x, np.log10(x))
a [1] [1].set_title('log')
plt.show()
```

```
import matplotlib.pyplot as plt
a1 = subplot2grid((3, 3),(0, 0), colspan = 2)
a2 = subplot2grid((3, 3),(0, 2), rowspan = 3)
a3 = subplot2grid((3, 3),(1, 0), rowspan = 3, colspan = 2)
import numpy as np
x = np.arange(1, 10)
a2.plot(x, x * x, 'rs--')
a2.set_title('square')
a1.plot(x, np.exp(x), 'go-')
a1.set_title('exp')
a3.plot(x, np.log(x), 'b>--')
a3.set_title('log')
plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt
data = [[30, 25, 50, 20],
[40, 23, 51,17],
[35, 22, 45, 19]]
X = np.arange(4)
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.bar(X + 0.00, data[0], color = 'b', width = 0.25)
ax.bar(X + 0.25, data[1], color = 'g', width = 0.25)
ax.bar(X + 0.50, data[2], color = 'r', width = 0.25)
```
